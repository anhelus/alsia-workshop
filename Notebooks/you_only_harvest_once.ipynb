### 2. Come usare un modello preaddestrato?

Per usare un modello preaddestrato, dovremo innanzitutto importare il file dei pesi (con estensione `pt`), e passarlo come argomento al costruttore della classe `YOLO`. In questo caso, stiamo utilizzando un modello di piccole dimensioni relativo all'ultima versione dell'architettura. Useremo il metodo predict per caricare tutte le immagini nella cartella `images` e vedere i risultati.

```python
from ultralytics import YOLO

model = YOLO('yolo26n.pt')
results = model.predict(source="images", show=True)
```

### 3. Uso pratico

L'uso pratico è possibile in molte situazioni. Ad esempio, potremo utilizzare i risultati restituiti per vedere quante bounding box sono state trovate nell'immagine.
