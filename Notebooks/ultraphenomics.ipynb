# Ultraphenomics



### 1. Organizzare i dati

La prima operazione da fare è quella di organizzare i dati. Una volta terminata la campagna di acquisizione, infatti, dovremo prendere i dati e deciderne la destinazione. Potremo:
1. Etichettarli
2. Utilizzarli con un metodo zero shot

Nel caso di etichettatura, dovremo utilizzare un software come Label Studio, con un po' di lavoro manuale. Nel caso si volesse utilizzare invece un metodo zero shot, il tutto è facilitato perché ci basterà andare ad utilizzare un prompt adeguato.

### 2. Effettuare l'addestramento

Una volta ottenuto il nostro dataset, potremo effettuare l'addestramento del modello. Per farlo, dovremo andare ad utilizzare il metodo `train` della libreria Ultralytics.

### 3. Interpretare i risultati

L'interpretazione dei risultati avviene mediante tutto ciò che emerge dopo l'addestramento. In particolare, potremo vedere sia i risultati in termini di metriche, sia quelli in termini pratici con le bounding box individuate.
